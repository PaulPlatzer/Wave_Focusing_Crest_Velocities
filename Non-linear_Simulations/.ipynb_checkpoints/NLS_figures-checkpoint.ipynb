```python
import numpy as np
import time
import matplotlib.pyplot as plt
```

```python
## Physical parameters ##
tp=12 #(s) peak period
grav=9.81 # (m/s^2) gravity acceleration at the earth's surface
w0=2*np.pi/tp # (s^-1) peak pulse
k0=(w0**2)/grav # (m^-1) peak wavenumber
lbda0=2*np.pi/k0 # (m) peak wavelength
c0=w0/k0
```

```python
f4=np.load('fig4bis.npz')
f4a=np.load('fig4bisAadim.npz')
```

```python
f4['A_RATIO'];f4['L_ESTIM'];f4['TFOC'];f4['TAU_ABS'];f4['COV'];f4['T_GRAD'];f4['C_GRAD'];f4['tau'];f4a['A_adim'];
```

```python
t_axis=f4a['t_axis'];A_adim=f4a['A_adim'];L_foc=f4a['L_foc']
```

```python
nstart=9
nend=len(f4['T_GRAD'][0])
N4=len(f4['T_GRAD'])
```

```python
T_GRAD=f4['T_GRAD']; TFOC=f4['TFOC']
```

```python
for l in range(N4):
    if len(f4['T_GRAD'][l])!=len(f4['T_GRAD'][0]):
        print('l='+str(l)+', length='+str(len(f4['T_GRAD'][l])))
```

```python
ntot=0
for l in range(N4):
    ntot+=len(f4['T_GRAD'][l])
```

```python
int(ntot-N4*(nstart+2))
```

```python
t_grad=np.zeros( int(ntot-N4*(nstart+2)) );tfoc=np.zeros( int(ntot-N4*(nstart+2)) )
A_ratio=np.zeros( int(ntot-N4*(nstart+2)) );L_estim=np.zeros( int(ntot-N4*(nstart+2)) )
tau_abs=np.zeros( int(ntot-N4*(nstart+2)) );c_grad=np.zeros( int(ntot-N4*(nstart+2)) )
i0=0; i1=len(f4['T_GRAD'][0])-nstart-2
t_grad[i0:i1]=f4['T_GRAD'][0][nstart+2:] ; tfoc[i0:i1]=f4['TFOC'][0][nstart+2:] 
A_ratio[i0:i1]=f4['A_RATIO'][0][nstart+2:] ; L_estim[i0:i1]=f4['L_ESTIM'][0][nstart+2:] 
tau_abs[i0:i1]=f4['TAU_ABS'][0][nstart+2:] ; c_grad[i0:i1]=f4['C_GRAD'][0][nstart+2:] 
for l in range(1,N4):
    i0+=len(f4['T_GRAD'][l-1])-nstart-2 ; i1+=len(f4['T_GRAD'][l])-nstart-2
    t_grad[i0:i1]=f4['T_GRAD'][l][nstart+2:]; tfoc[i0:i1]=f4['TFOC'][l][nstart+2:];
    A_ratio[i0:i1]=f4['A_RATIO'][l][nstart+2:]; L_estim[i0:i1]=f4['L_ESTIM'][l][nstart+2:];
    tau_abs[i0:i1]=f4['TAU_ABS'][l][nstart+2:]; c_grad[i0:i1]=f4['C_GRAD'][l][nstart+2:];
```

```python
ind=np.argsort(t_grad)
t_grad=t_grad[ind]; tfoc=tfoc[ind]; A_ratio=A_ratio[ind]; L_estim=L_estim[ind]; tau_abs=tau_abs[ind]; c_grad=c_grad[ind]; 
```

```python
import pandas
```

```python
ns_roll=100
```

```python
len(pandas.rolling_median(tfoc,ns_roll,min_periods=int(ns_roll/2-2),center=True))
```

```python
rolmedian_tfoc=pandas.rolling_median(tfoc,ns_roll,min_periods=int(ns_roll/2-2),center=True)
```

```python
rol95per_tfoc=pandas.rolling_quantile(tfoc,ns_roll,.95,min_periods=int(ns_roll/2-2),center=True)
rol05per_tfoc=pandas.rolling_quantile(tfoc,ns_roll,.05,min_periods=int(ns_roll/2-2),center=True)
```

```python
plt.clf(); fig=plt.figure(); fig.add_axes(); ax=fig.add_subplot(111)
ax.plot(np.array([-60,10]),np.zeros(2),'k',label=r'$t_{foc}/T_p=0$')
ax.plot(t_grad*w0/(2*np.pi),rolmedian_tfoc*c0*k0/(2*np.pi),'-k',linewidth=1,label='Estimation')
ax.fill_between(t_grad*w0/(2*np.pi),rol05per_tfoc*c0*k0/(2*np.pi),rol95per_tfoc*c0*k0/(2*np.pi),color='cyan')
ax.set_title('Focusing time estimation',fontsize=16)
ax.set_xlabel(r'$t_{end}/T_p$',fontsize=16)
ax.set_ylabel('$t_{foc}/T_p$',fontsize=16)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.,fontsize=16)
ax.tick_params(length=7,labelsize=16,direction='out',pad=10)
ax.set_ylim([-5,5])
ax.set_xlim([-30,0])
fig.set_size_inches([7,5])
plt.show()
```
